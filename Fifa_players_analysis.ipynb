### 1(a). Count all the players and sorting them in descending order

```python
# Type your answers here #
# You may use multiple cells, but make sure you answer # 
# the questions using the cell above #
df_player_count = df.groupby(['Nationality']).size().reset_index(name='counts').sort_values(by='counts',ascending=False)
df_player_count
```

### 1(b). Get the list of nations whose players are atleast 50 in count

```python
nation_list = []
df_player_count = df_player_count.loc[df_player_count['counts']>=50]
nation_list = df_player_count.values
nation_list
```

### 1(c). Get the overall mean of each nation whose player count is atleast 50. Then sort them in descending order

```python
pd.options.mode.chained_assignment = None  # default='warn'
overall_array = []
for nation, count in nation_list:
    overall = df.loc[df["Nationality"]==nation]["Overall"].mean()
    overall_array.append(overall)
df_player_count["OverallAvg"] = overall_array
df_player_count = df_player_count.sort_values(by='OverallAvg', ascending=False)
df_player_count.head(5)
```

### 2(a). Get the list of all clubs with their player counts

```python
df_clubs = df.groupby(['Club'])['WageEUR'].sum().reset_index(name="SUMEUR").sort_values(by="SUMEUR", ascending=False)
df_clubs
```

### 2(b). Get the sum of wages by of players from each club and then sort them in descending order according to their wage sum

```python
clubs = df_clubs.values
wages = []
for club, count in clubs:
    club_wage = df.loc[df["Club"]==club]["WageEUR"].sum()
    wages.append(club_wage)
df_clubs["WageSum"] = wages
```

```python
df_clubs = df_clubs.sort_values(by="WageSum", ascending=False)
df_clubs.head()
```

### 3. Get the players with age less than or equal to 19, then get the mean of their growth and overall performance. Finally represent them in a new dataframe to analyse better

```python
df_young = df.loc[df["Age"]<=19]
df_young = df_young.sort_values(by="Age",ascending=False)
df_potential = df_young.groupby(['Age'])['Growth'].mean()
df_young_overall = df_young.groupby(['Age'])['Overall'].mean()

new_df_young = pd.DataFrame(df_potential).join(df_young_overall)
new_df_young
```
